```python
#Matrix_5x5
from collections import*
chars='abcdefghijklmnopqrstuvwxyz'
chars=chars.upper()
key=input('Enter the key: ')
key=key.upper()
chars=chars.replace('J','I')
key=key.replace('J','I')
nv=key+chars
print(nv)
kpchaf=''.join(OrderedDict.fromkeys(nv))
print(kpchaf)

matrix=[[0 for i in range(5)]for j in range(5)]
index=0
for i in range(5):
    for j in range(5):
        matrix[i][j]=kpchaf[index]
        index+=1

#plain text
plain_txt=input('Enter the plain text: ')
plain_txt=plain_txt.upper()
plain_txt=plain_txt.replace('J','I')
for t in range(0,len(plain_txt),2):
    ch1=plain_txt[t]
    ch2=plain_txt[t+1]
    if ch1==ch2:
        plain_txt=plain_txt[0:t+1]+'X'+plain_txt[t+1:]
if(len(plain_txt)%2)!=0:
    plain_txt=plain_txt+'X'    
print(plain_txt) 

#Encrption
chtext=''
for t in range(0,len(plain_txt),2):
    ch1=plain_txt[t]
    ch2=plain_txt[t+1]
    for i in range(5):
        for j in range(5):
            if ch1==matrix[i][j]:
                r1=i
                c1=j
            if ch2==matrix[i][j]:
                r2=i
                c2=j
    if r1==r2:
        nc1=(c1+1)%5
        nc2=(c2+1)%5
        chtext+=matrix[r1][nc1]
        chtext+=matrix[r1][nc2]
    elif c1==c2:
        nr1=(r1+1)%5
        nr2=(r2+1)%5
        chtext+=matrix[nr1][c1]
        chtext+=matrix[nr2][c2]
    else:
        chtext+=matrix[r1][c2]
        chtext+=matrix[r2][c1]
print(chtext)

#Decrption 
ptext=''
for t in range(0,len(chtext),2):
    ch1=chtext[t]
    ch2=chtext[t+1]
    for i in range(5):
        for j in range(5):
            if ch1==matrix[i][j]:
                r1=i
                c1=j
            if ch2==matrix[i][j]:
                r2=i
                c2=j
    if r1==r2:
        nc1=(c1-1)%5
        nc2=(c2-1)%5
        ptext+=matrix[r1][nc1]
        ptext+=matrix[r1][nc2]
    elif c1==c2:
        nr1=(r1-1)%5
        nr2=(r2-1)%5
        ptext+=matrix[nr1][c1]
        ptext+=matrix[nr2][c2]
    else:
        ptext+=matrix[r1][c2]
        ptext+=matrix[r2][c1]
print(ptext) 
```
